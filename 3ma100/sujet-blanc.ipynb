# Exercise 1: The golden ratio

<!-- Exercice proposé par Cindy Guichard, novembre 2018 -->
We would like to approximate the golden ratio $\varphi = \dfrac{1+\sqrt{5}}{2}$.

First we define a sequence $(F_n)_{n\geqslant0}$ with $F_0=F_1=1$ and $F_{n}=F_{n-1}+F_{n-2}$ for $n\ge 2$. We then obtain the following results:

$$
\lim_{n \rightarrow +\infty} \dfrac{F_{n+1}}{F_{n}} = \varphi.
$$

**a)**
Write a function `Fibo(epsilon)` which given a precision $\varepsilon$ returns the smallest $n$ so that $\left| \varphi -  \dfrac{F_{n+1}}{F_{n}} \right| < \varepsilon$.



```python
from math import sqrt

golden = (1 + sqrt(5))/2

def Fibo(epsilon):
    n = 0
    F0, F1 = 1, 1
    while abs(golden - F1/F0) >= epsilon:
        F0, F1 = F1, F0 + F1
        n += 1
    return n
```


We admit also that $\varphi$ is the only positive solution for $x^2-x-1=0$.

**b)**
Write another function `Newton(epsilon, x0)` which given $\varepsilon$ returns the smallest $k$ so that $\left| \varphi - x_k \right| < \varepsilon$ where $x_k$ is determined by the Newton's method for $f(x)=x^2-x-1$, in which we have:

$$
x_{k+1} = x_k - \dfrac{f(x_k)}{f^\prime(x_k)} \,, k \ge 0.
$$

We are going to choose $x_0=3$.



```python
def Newton(epsilon, x0):
    f =  lambda x: x**2 - x - 1
    df = lambda x: 2*x - 1
    k = 0
    while abs(golden - x0) >= epsilon:
        x0 -= f(x0)/df(x0)
        k += 1
    return k
```


**c)**
For $\varepsilon = 10^{-i}$ and $i \in \{2, 4, 6, 8\}$ compare the number of iterations needed between the 2 proposed methods to approximate $\varphi$ until the given precision. Which one seems more *efficient*?



```python
i_ = (2, 4, 6, 8)
x0 = 3
for i in i_:
    epsilon = 10**-i
    print(f"i = {i}. Number of iterations:")
    print(f"- Fibo:   {Fibo(epsilon)}")
    print(f"- Newton: {Newton(epsilon, x0)}")
    print("")
```

> __Newton's__ method seems to be way more efficient. It requires much less iterations to approximate below the small precision.


**d)**
What if we initialize Newton with $x_0=0$ ?

```python
Newton(10e-5, 0)
```

> The iteration runs forever since it cannot approach the answer. Choice of $x_0$ matters.

# Exercise 2: Least squares method

<!-- Exercice proposé par Marie Postel, novembre 2018 -->
The file `population.txt` contains the statistical data of population in some French cities, ordered by sex and age. Some first few lines include:

        CODGEO;LIBGEO;SEXE;AGED100;NB
        13201;Marseille 1er Arrondissement;1;000;293.98064823
        13201;Marseille 1er Arrondissement;1;001;311.40411923
        13201;Marseille 1er Arrondissement;1;002;257.55156994
        13201;Marseille 1er Arrondissement;1;003;249.26498077


The first column is postal codes, the second is city's names, the third is sexes (1 for male, 2 for female), the fourth is ages, and the fifth is population counts. For example, the second line shows that there are 293.98 persons, male, age of 0 in the 1st arrondissement of Marseille.

**a)**
Write a function `pop(code)` returning a list or a Numpy vector the population by age without distinction between male and female in the city with the postal code `code`. Use this function to determine the total population in 13201 and 75102.

```python
import pandas as pd  # Pandas is easy to use to process data from files like this

df = pd.read_csv("population.txt", sep=";")
df.head()
```

```python
# Since we require that the function returns the population without differentiating sexes, we can achieve this by grouping and calculating
# the sum of population by age.

def pop(code):
    df_code = df.loc[df['CODGEO']==code]  # filtering out only rows with CODGEO that we look for
    pop_series = df_code.groupby(["AGED100"]).sum()['NB']  # return a Pandas series of sums of population by age
    return pop_series.to_numpy()

codes = [13201, 75102]

for code in codes:
    print(f"Total population in {code} is {np.sum(pop(code))}")
```


**b)**
Write a function `plot_pop(code)` plotting the proportions of the population (normalized by the total number) by age without distinction between male and female in the city with the postal code `code`. Test for 13201 and 75102.



```python
def plot_pop(code):
    Y = pop(code)
    X = range(len(Y))
    total = np.sum(Y)
    geo_name = df.loc[df['CODGEO']==code]['LIBGEO'].iloc[0]
    
    plt.subplots(figsize=(20,8))
    plt.plot(X, Y/total, label="Real population")
    plt.xlabel("Age")
    plt.ylabel("Proportion to total population")
    plt.title(f"Proportion of different ages to total population in {geo_name}")
```

```python
plot_pop(13201)
plt.show()
```

```python
plot_pop(75102)
plt.show()
```


For a given city, by noting $p(a)$ the proportion of the population having age $a$, we search for the coefficients $r_0$, $r_1$, $r_2$ so that the probability distribution

$$
p(a) = r_0+r_1a+r_2a^2
$$

fits the best for the age groups $a>25$. In order to do so, we solve for the least squares problem

$$
\min_{\vec{r}\in \mathbb{R}^3}||X\vec{r}-\vec{p}||^2
$$

in which by denoting the age vector $\vec{a}=(25,26,\dots,100)$, $X$ is the $76\times 3$ matrix where $X_{i,1}=1$, $X_{i,2}=a_i$, $X_{i,3}=a_i^2$ and $\vec{p}$ is the vector representing the population by age $p_i = p(a_i)$. The solution for this problem is $\vec{r} = (r_0,r_1,r_2)\in \mathbb{R}^3 $. This solution would also satisfy the equation

$$
~^t\!XX\vec{r}=~^t\!X\vec{p}
$$

where $~^t\!X$ designates the transpose of $X$. For the postal codes 13201 and 75102, answer the following questions:

**c)**
Create the matrix $X$ and the vector $\vec{p}$.



```python
def Xp(code):
    X = np.ones((76,3), dtype=int)
    X[:, 1] = np.arange(25,101)
    X[:, 2] = X[:, 1]**2
    
    p = pop(code)
    p = (p/np.sum(p))[25:]
    return X, p
```


**d)**
Determine the solution $\vec{r}$.



```python
# Using the function np.linalg.solve to solve for x in Ax=b, where A=tX.X and b=tX.p

for code in codes:
    X, p = Xp(code)
    r = np.linalg.solve(X.T @ X, X.T @ p)
    print(f"Solution for {code} is {r}")
```


**e)**
Plot the theoretical curve $r_0+r_1a+r_2a^2$ by age on top of the collected data.

```python
theory = lambda a, r: r[0] + r[1]*a + r[2]*a**2
x = np.arange(25, 101)

for code in codes:
    X, p = Xp(code)
    r = np.linalg.solve(X.T @ X, X.T @ p)
    plot_pop(code)
    plt.plot(x, theory(x, r), label="Theoretical population")
    plt.legend()
    plt.show()
```


Finally, we are going to implement our very own classifier: with a new image of hand-written digit, we predict the label with which the image resembles. To achieve this, we first have to divide our database into two parts of similar size: the first part to train the classifier (or model) `X_train` and `y_train`, the second to test after the model has been trained `X_test` and `y_test`.

**g)**

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5)
```


**h)**
Using the training set, calculate the centroids of labels from 0 to 9, then save the results in a variable `centroids_train`.

```python
centroids_train = []

for i in range(10):
    mask = y_train==i
    mean = np.mean(X_train[mask], axis=0)
    centroids_train.append(mean)
```


**i)**
For each digit in the testing set `X_test`, determine which centroid the digits belongs to in `centroids_train` by calculating the euclidean norm.

```python
from scipy.spatial.distance import euclidean

centroids_predict = []
for digit in X_test:
    predict = min(range(10), key=lambda i: euclidean(centroids_train[i], digit))
    centroids_predict.append(predict)
```
